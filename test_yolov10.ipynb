```python
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

```python
IMG_PATH = './images/HCMC_Street.jpg'
result = model(source = IMG_PATH)[0]
```

```python
from pytube import YouTube
YOUTUBE_VIDEO_PATH = 'https://youtu.be/wqPSsu7XQ74'
DOWNLOAD_PATH = '/path/to/downloaded/video.mp4'

# Download the video
yt = YouTube(YOUTUBE_VIDEO_PATH)
stream = yt.streams.filter(file_extension='mp4').first()
stream.download(output_path=DOWNLOAD_PATH)

# Now use the downloaded video path
video_result = model(source=DOWNLOAD_PATH)
```

```python
for i, result in enumerate(video_result):
    result.save(f'./images/video/video_frame_{i}_predict.png')  # Save each frame with a unique name
```
